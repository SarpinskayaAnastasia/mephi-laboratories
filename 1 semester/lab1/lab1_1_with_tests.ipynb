# Задание 1
Написать тело функции `check_if_pow_of_m`, принимающей в качестве аргумента числа `n` и `m` (`m`,`n`>=0), проверяющей, является ли `n` степенью `m` (`n`=`m`^`x`)(без использования функций извлечения корня и вычисления логарифма).В ответе программа должна возвращать булево значение (`True`, `False`).

Пример: задано числа `256` и `2`

Ответ: `True`


```python
def check_if_pow_of_m(n,m):
  ans = False
  x = 0
  while not ans:
    if n == m ** x:
      ans = True
    elif (n < m ** x and m > 1) or (n > m ** x and m < 1):
      break
    else:
      x += 1

  return ans
```

```python
test_data = [[[0.25, 0.5], True], [[0, 2], False], [[1, 2], True], [[2, 2], True], [[3, 2], False], [[4, 2], True], [[4, 3], False], [[5, 6], False], [[6, 3], False], [[7, 2], False], [[8, 2], True], [[8, 3], False], [[16, 4], True], [[16, 3], False], [[16, 2], True], [[256, 2], True]]
for i in test_data:
    x = check_if_pow_of_m(i[0][0],i[0][1])
    assert ( x== i[1]), f"Test with {i[0]} failed, got {x} expected {i[1]}"
print("ALL TESTS PASSED!")
```

# Задание 2
Написать тело функции `pascal_triangle_sum`, принимающей параметр `n` - целое число (n>=0). Функция должна возвращать сумму всех элементов всех слоев треугольника Паскаля высоты `n`.

Треугольник Паскаля - бесконечная таблица, в которой на вершине и по бокам расположены единицы. Любое другое число равно сумме двух расположенных над ним чисел.
![Треугольник Паскаля](https://avatars.dzeninfra.ru/get-zen_doc/27036/pub_5f7212ba315c38725cd0f735_5f72133eba91764cb5f30e5f/scale_1200)

Пример:
`n` = 3

Ответ: `7`

```python
def pascal_triangle_sum(n):
  s = 0
  for i in range(1, n+1):
    C = 1
    for j in range(1, i+1):
      s += C
      C = C * (i - j) // j

  return s
```

```python
test_data = [[0, 0], [1, 1], [2, 3], [3, 7], [4, 15], [5, 31]]
for i in test_data:
    x = pascal_triangle_sum(i[0])
    assert ( x== i[1]), f"Test with {i[0]} failed, got {x} expected {i[1]}"
print("ALL TESTS PASSED!")
```

# Задание 3
Напишите тело функции, находящей в заданном списке длины `n` (n >= 0) индексы всех элементов, которые меньше своих соседей. В случае нехватки количества соседей считать не хватающий элемент большим, чем рассматриваемый. Для крайних элементов (первый, последний) соседними считаются действительный сосед и элемент с противоположного конца списка (например, дан список `[1,2,…,8,9]`. Соседями числа 1 считается 2 и 9, а соседями числа 9 – 1 и 8).
Для возврата значения используйте команду `return <data>`

Пример: задан список `[1,5,6,9,3,5,4]`

Вывод программы: `[0,4]`

```python
def less_than_neighbours(a):
  b = []
  if not a:
    return []
  elif len(a) == 1:
    return [0]
  elif len(a) == 2:
    if a[0] > a[1]:
      return [1]
    elif a[0] < a[1]:
      return [0]
    else:
      return []
  else:
    if a[-1] > a[0] < a[1]:
      b += [0]
    for i in range(1, len(a) - 1):
      if a[i - 1] > a[i] < a[i + 1]:
        b += [i]
    if a[0] > a[-1] < a[-2]:
      b += [len(a) - 1]
  return b
```

Напишите тело функции `delete_extra_element`, принимающей в качестве параметра `a` список натуральных чисел с парами дублирующихся элементов, и одним элементом без дубликата. Длина списка составляет `n` (`n`>=0). Необходимо удалить лишний элемент. При этом в основном списке гарантированно не повторяется не более 1 элемента. В ответе необходимо указать исправленный список, значение лишнего элемента и индекс удаленного элемента. В случае, если никакой элемент не был удален, в качестве значения элемента и индекса указать значение -1.

Пример: `[1,2,3,1,2,3,5]`.

Ответ: `([1,2,3,1,2,3], 5, 6)`

```python
from re import I
def delete_extra_element(a):
  if not a:
    return [], -1, -1
  v = -1
  i = -1
  for ind in range(len(a)):
    if a.count(a[ind]) == 1:
      v = a[ind]
      i = ind
      break
  if i != -1:
    del a[i]

  return a, v, i
```

# Задание 5
Написать тело функции `empty_the_list`, принимающей в качестве аргумента список `a` уникальных целых чисел длины `n` (`n`>=0). Со списком можно производить две операции:
- Если первый элемент – минимальный элемент списка – удалить его из списка
- иначе – перенести первый элемент в конец списка.

Необходимо вернуть количество операций, требующихся для полной очистки списка.

Пример: Задан список `[3,4,-1]`

Ответ программы: 5

```python
def emtpy_the_list(a):
  if not a:
    return 0
  count = 0
  while a:
    if a[0] == min(a):
      del a[0]
      count += 1
    else:
      a = a[1:] + [a[0]]
      count += 1

  return count
```

```python
test_data = [[[3, 4, -1], 5], [[1, 2, 4, 3], 5], [[1, 2, 3], 3], [[], 0], [[1], 1]]
for i in test_data:
    assert(i[1] == emtpy_the_list(i[0][:])), f"Test with {i[0]} failed, got {emtpy_the_list(i[0][:])} expected {i[1]}"
print("ALL TESTS PASSED!")
```
